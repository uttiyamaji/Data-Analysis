Analysing the restaurents in Delhi,  It's more about understanding how price varies with cuisine

```python
import numpy as np
import pandas as pd
```

```python
# directly downloading from github
file_path = "https://raw.githubusercontent.com/uttiyamaji/Data-Analysis/master/Zomato_restaurents_data_analysis/zomato.csv"

# contains the whole df
df = pd.read_csv(file_path, encoding = 'latin')

# only the ones in Delhi
df_delhi = df.copy()[df['City'] == 'New Delhi']

df_delhi.head()
```

Many of the variables are not needed, we might drop them at this point

Pre-precessing the cuisines in order to analyse them

```python
cuisines = df_delhi.Cuisines

cuisines_list = []

for word in cuisines:
    cuisines_list.append(word.split(','))
    
cuisines_list
```

```python
cuisine_corpus = []

for x in cuisines:
    xs = x.split(',')
    cuisine_corpus.extend(xs)

cuisine_corpus = [x.strip() for x in cuisine_corpus]

cuisine_corpus_set = set(cuisine_corpus)

cuisine_dict = {ix:name for ix,name in enumerate(cuisine_corpus_set)}
cuisine_dict_reverse = {name:ix for ix,name in enumerate(cuisine_corpus_set)}
```

```python
ids = []
for cui in cuisines_list:
    cui_list = []
    for cuis in cui:
        cuis = cuis.strip()
        idx = list(cuisine_dict.values()).index(cuis)
        #print(idx)
        cui_list.append(idx)
        #print(cui_list)
    ids.append(cui_list)

    
print(ids)
```

```python
# lists of list
one_hc = []

#
for x in ids:
    enc = [0]*len(cuisine_dict.values())
    for y in x:
        enc[y] = 1
    one_hc.append(enc)

one_hc
```

```python
cuisine_extracted = np.array(one_hc)
cuisine_extracted.shape


cuisine_vars = pd.DataFrame(cuisine_extracted, columns = cuisine_dict_reverse.keys())



df_delhi_cuisine_price = pd.concat([cuisine_vars.reset_index(drop = True), 
                                    df_delhi['Average Cost for two'].reset_index(drop = True),
                                    df_delhi['Aggregate rating'].reset_index(drop = True)], axis = 1)
#df_delhi_cuisine_price.reset_index(drop = True)
df_delhi_cuisine_price.head()
                                
```
